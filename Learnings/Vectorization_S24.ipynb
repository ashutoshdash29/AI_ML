BAG OF WORDS

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
CV = CountVectorizer()
```

```python
data = [
    'i am a boy and i am also a man but i am not a girl or a woman',
    'she did not know what she is about to do but i think i know what she is going to do',
    'do you have any idea how much i have worked to be this successful today',
    'only blabbering will take you nowhere i dod not reach where i am today by blabbering'
    'hope this many words that i wrote are enough to give the context i am trying to set'
    ]
```

```python
count_vec = CV.fit_transform(data)
```

```python
type(count_vec)
```

```python
BagOfWords = pd.DataFrame(count_vec.toarray(), columns = CV.get_feature_names_out())
```
