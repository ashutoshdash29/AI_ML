LDA Latent Dirichlet Allocation

```python
import pandas as pd
from pathlib import Path
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import gensim
import gensim.corpora as corpora
```

```python
parent_dir = Path.cwd().parent
data = pd.read_csv(parent_dir.joinpath('Supporting_Docs').joinpath('news_articles.csv'))
```

```python
data.info()
```

```python
articles = data['content']
```

```python
articles = articles.str.lower().apply(lambda x: re.sub(r"[^\w\s]","",x))
```

```python

# removing stopwords
en_stopwords = stopwords.words('english')
articles = articles.apply(lambda x : " ".join([word for word in x.split(" ") if word not in en_stopwords]))
```

```python
#tokenize
articles = articles.apply(lambda x : word_tokenize(x))

```

```python
#Stemming
ps = PorterStemmer()
articles = articles.apply(lambda x : [ps.stem(y) for y in x])
```

```python
articles
```

```python
dictionary = corpora.Dictionary(articles)
```

```python
print(dictionary)
```

```python
dict(list(dictionary.items())[:10])
```

```python
doc_term_matrix = [dictionary.doc2bow(article) for article in articles]
```

```python
num_of_topics = 5
```

```python
LDA_model = gensim.models.LdaModel(doc_term_matrix, id2word=dictionary, num_topics=num_of_topics)
```

```python
LDA_model.print_topics(num_topics=num_of_topics, num_words=10)
```

LSA Latent Schemantic Analysis

```python
from gensim.models import LsiModel
from gensim.models.coherencemodel import CoherenceModel
import matplotlib.pyplot as plt
import numpy as np
```

```python
LSI_Model = LsiModel(doc_term_matrix, id2word=dictionary, num_topics=num_of_topics)
```

```python
LSI_Model.print_topics(num_topics=num_of_topics, num_words=10)
```

```python
coherence_value  = []
model_list = []

for i in range(2,12):
    model = LsiModel(doc_term_matrix, id2word=dictionary, num_topics=i)
    coherence_model = CoherenceModel(model, texts=articles, dictionary=dictionary, coherence='c_v')
    model_list.append(model)
    coherence_value.append(coherence_model.get_coherence())
    print(coherence_model.get_coherence())
```

```python
plt.plot(range(2,12),coherence_value)
plt.xlabel(xlabel='Number of Topics',loc='center')
plt.ylabel(ylabel='Coherence Value',loc='center')
plt.figaspect(float(9/16))
plt.legend(('Cohereence Value'), loc='best')
plt.grid()
plt.xticks(np.arange(0,15,step=1))
plt.show()
```
