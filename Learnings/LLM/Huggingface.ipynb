Huggingface Transformers

```python
from transformers import pipeline
```

```python
sentiment_classifier = pipeline('sentiment-analysis')
```

```python
sentiment_classifier("I am Excited today")
```

Giving a model to pipeline explicitly

```python
ner = pipeline("ner",model="dslim/bert-base-NER")
```

```python
ner("My name is Ashutosh Dash and I want to get into Accenture for the role of Python Fullstack Developer")
```

Zero Shot classifier from Facebook

```python
classifier = pipeline("zero-shot-classification",model="facebook/bart-large-mnli")
```

```python
sequence_to_classify = "Ranveer Singh is the best Actor in the whole Universe"
candidate_labels = ['Actor', 'Movie', 'Cosmos', 'People', 'Industry']
classifier(sequence_to_classify, candidate_labels)
```

```python
classifier(sequence_to_classify, candidate_labels, multi_label=True)
```
